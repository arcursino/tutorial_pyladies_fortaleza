Muitos datas est√£o sem o ano, ent√£o vamos filtrar apenas as ladies que tem o campo data completo

```python
idade = df[df['data_anivers√°rio'].apply(lambda x: len(x) > 6) ]
```

```python
from datetime import datetime
ano_atual = datetime.now().year
```

```python
idade['data_anivers√°rio'].info()
```

Est√° como objeto ent√£o vamos transformar para datetime

```python
idade['niver'] = pd.to_datetime(idade['data_anivers√°rio'], format='mixed')
```

Extrair o ano da coluna niver

```python
idade['ano_nascimento'] = idade['niver'].dt.year
```

```python
idade['idade_lady'] = ano_atual - idade['ano_nascimento']
```

```python
# qual a idade das ladies por ordem crescente
idade['idade_lady'].sort_values()
```

```python
# qual idade da lady mais nova?
idade['idade_lady'].min()
```

```python
# qual a m√©dia das idades das ladies
idade['idade_lady'].mean()
```
