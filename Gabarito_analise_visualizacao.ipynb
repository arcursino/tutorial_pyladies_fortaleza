### Desafio - Nível Hard
- Qual a idade das ladies?
- Qual a lady mais nova?
- Qual a média da idade das ladies?

```python
df.columns
```

```python
df['niver'] = pd.to_datetime(df['data_aniversário'], format='%d/%m/%Y')
```

```python
#Como as datas estão formatada?
df['data_aniversário'].unique()
```

Muitos datas estão sem o ano, então vamos filtrar apenas as ladies que tem o campo data completo

```python
idade = df[df['data_aniversário'].apply(lambda x: len(x) > 6) ]
```

```python
from datetime import datetime
ano_atual = datetime.now().year
```

```python
idade['data_aniversário'].info()
```

Está como objeto então vamos transformar para datetime

```python
idade['niver'] = pd.to_datetime(idade['data_aniversário'], format='mixed')
```

Extrair o ano da coluna niver

```python
idade['ano_nascimento'] = idade['niver'].dt.year
```

```python
idade['idade_lady'] = ano_atual - idade['ano_nascimento']
```

```python
# qual a idade das ladies por ordem crescente
idade['idade_lady'].sort_values()
```

```python
# qual idade da lady mais nova?
idade['idade_lady'].min()
```

```python
# qual a média das idades das ladies
idade['idade_lady'].mean()
```
